```python
import numpy as np
import gensim
import os
import re

from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from gensim import corpora

from gensim.models.ldamulticore import LdaMulticore

import pandas as pd
```

```python
path = './data/Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products_May19.csv'
```

```python
amazon = pd.read_csv('./data/Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products_May19.csv')
```

```python
amazon.head()
```

```python
# 1) Plain Python - ''.split command
# 2) Spacy - just the lemmas from the document
# 3) Gensim - simple_preprocess

def tokenize(text):
    "Complete this function"
    
    return [token for token in simple_preprocess(text) if token in STOPWORDS]
```

```python
amazon["tokens"] = amazon["reviews.text"].apply(tokenize)
```

```python
amazon.head()
```

```python
# A Dictionary Representation of all the words in our corpus
id2word = corpora.Dictionary(amazon["tokens"])
```

```python
id2word.token2id['name']
```

```python
id2word[25]
```

```python
len(id2word.keys())
```

```python
# Let's remove extreme values from the dataset
id2word.filter_extremes(no_below=5, no_above=0.95)
```

```python
len(id2word.keys())
```

```python
corpus = [id2word.doc2bow(text) for text in amazon["tokens"]]
```

```python
corpus[48][:10]
```

```python
lda = LdaMulticore(corpus=corpus,
                   id2word=id2word,
                   random_state=723812,
                   num_topics = 10,
                   passes=10,
                   workers=10
                  )
```

```python
lda.print_topics()
```

```python
words = [re.findall(r'"([^"]*)"',t[1]) for t in lda.print_topics()]
```

```python
topics = [' '.join(t[0:5]) for t in words]
```

```python
for id, t in enumerate(topics): 
    print(f"------ Topic {id} ------")
    print(t, end="\n\n")
```

```python
from gensim.models.coherencemodel import CoherenceModel

def compute_coherence_values(dictionary, corpus, limit, start=2, step=3, passes=5):
    """
    Compute c_v coherence for various number of topics

    Parameters:
    ----------
    dictionary : Gensim dictionary
    corpus : Gensim corpus
    limit : Max num of topics
    passes: the number of times the entire lda model & coherence values are calculated

    Returns:
    -------
    coherence_values : Coherence values corresponding to the LDA model with respective number of topics
    """
    
    coherence_values = []
    
    for iter_ in range(passes):
        for num_topics in range(start, limit, step):
            model = LdaMulticore(corpus=corpus, num_topics=num_topics, id2word=dictionary, workers=8)
            coherencemodel = CoherenceModel(model=model,dictionary=dictionary,corpus=corpus, coherence='u_mass')
            coherence_values.append({'pass': iter_, 
                                     'num_topics': num_topics, 
                                     'coherence_score': coherencemodel.get_coherence()
                                    })

    return coherence_values
```

```python
# Can take a long time to run.
coherence_values = compute_coherence_values(dictionary=id2word, 
                                                        corpus=corpus,
                                                        start=2, 
                                                        limit=40, 
                                                        step=6,
                                                        passes=40)
```
